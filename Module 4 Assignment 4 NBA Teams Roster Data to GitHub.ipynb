```python
# Import libraries
import pandas as pd

# Import Basketball Reference roster tables for 2023-24 Knicks, Raptors, Thunder, Clippers
nyk_roster_2023_24 = pd.read_csv("datamod4/knicks2023-24roster.csv", index_col="Player")
tor_roster_2023_24 = pd.read_csv("datamod4/raptors2023-24roster.csv", index_col="Player")
okc_roster_2023_24 = pd.read_csv("datamod4/thunder2023-24roster.csv", index_col="Player")
lac_roster_2023_24 = pd.read_csv("datamod4/clippers2023-24roster.csv", index_col="Player")
```

```python
# Bind rows the of 4 tables and add nationality column label
nyk_tor_okc_lac_2023_24_rosters = pd.concat([nyk_roster_2023_24, tor_roster_2023_24, okc_roster_2023_24, lac_roster_2023_24])
nyk_tor_okc_lac_2023_24_rosters = nyk_tor_okc_lac_2023_24_rosters.rename(columns={"Unnamed: 6":"Nationality"})
nyk_tor_okc_lac_2023_24_rosters
```

```python
nyk_tor_okc_lac_2023_24_rosters.to_csv("datamod4/nyk_tor_okc_lac_2023-24_rosters.csv", na_rep="NULL")
```
